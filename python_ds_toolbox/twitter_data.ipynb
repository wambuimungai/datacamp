```python
#importing pandas library to access data
import pandas as pd

#importing twitter data from the csv file

twitter_data =  pd.read_csv('tweets.csv')

twitter_data.head()
list(twitter_data.columns)
```

```python
#use function
def  count_entries(df, col_name):
    
    #initialise an empty dictionary. langs_count dictionary to show count of each language used in tweets
    langs_count = {}
    
    try:
        col = twitter_data[col_name]

        #iterate through the lang column. if lang already in the dictionary, add 1 else add the lang 
        for entry in col:

            if entry in langs_count.keys():
                langs_count[entry] += 1
            else:
                langs_count[entry] = 1

        return langs_count   
    except:
        print('The DataFrame does not have a '+ col_name + 'column')
    
result = count_entries(twitter_data, 'lang')
              
print(result)        
```
